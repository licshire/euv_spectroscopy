## After manual labelling different runs are stored in different .csv files here we merge these into one

```python
import os

import pandas as pd
```

```python
x = 0
for element in os.listdir('.'):
    if '.csv' in element:
        if x == 0:
            merge_df = pd.read_csv(element)
            x+=1
            continue
        current_df = pd.read_csv(element)
        merge_df = pd.concat([merge_df,current_df], ignore_index=True)

merge_df.head()
```

Preview after merge there are 5 columns with no values at all (Note,Calibration1,Calibration2,Calibration3,Calibration4) although we will need these in the future for now I'll take them out. Also forgot to tell pandas not to use index column so taking that out as well.

```python
merge_df = merge_df.iloc[:,1:-5]
merge_df.head()
```

For 962 files the code couldn't extract the element. Most of these are background but some are just formats the code couldn't read. Looping through these and adding them manually.

```python
def checkZeros(df):
    
    for i, element in enumerate(df.Element):
        if element == 0:
            eid = input(df.FileName[i] + ': ')
            if eid == '':
                eid = 0
            else:
                eid = int(eid)
            df.Element.loc[i] = eid
            
    return df
```

```python
merge_df = checkZeros(merge_df)
```
