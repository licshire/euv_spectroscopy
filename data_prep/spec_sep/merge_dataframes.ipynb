```python
import os

import pandas as pd
```

```python
x = 0
for element in os.listdir('.'):
    if '.csv' in element:
        if x == 0:
            merge_df = pd.read_csv(element)
            x+=1
            continue
        current_df = pd.read_csv(element)
        merge_df = pd.concat([merge_df,current_df], ignore_index=True)

merge_df.head()
```

```python
merge_df = merge_df.iloc[:,1:-5]
```

```python
merge_df.to_csv('specBase.csv', index=None)
```
