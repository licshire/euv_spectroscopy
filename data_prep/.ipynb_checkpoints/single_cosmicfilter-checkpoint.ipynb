```python
import sys
sys.path.append('./cosmic')

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

import scipy_spe as spe
```

```python
arr = spe.load('cosmic/EUV053_150p1mA_1D15_4p10KV_60s_15sdump_Mo_09162015.SPE')
arr.shape # (15, 2048, 1)
arr = arr.reshape(arr.shape[:-1]).T
```

```python
arr_avg = np.average(arr, axis=1).reshape(2048,1) # average of every line
arr_std = np.std(arr, axis=1).reshape(2048,1)
c_mask = arr-arr_avg # shift to 0
np.where(c_mask>-arr_std*3)
```

```python
igorcosmic = np.loadtxt('cosmic/EUV53_cosmic.txt')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(15,15))
ax1.plot(igorcosmic, linewidth=.5)
ax2.plot(c_mask, linewidth=.5)
```
